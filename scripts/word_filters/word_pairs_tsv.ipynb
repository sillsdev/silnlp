```python
input_file = "word_pairs_sorted.txt"
output_file = "word_pairs_sorted.tsv"

with open(input_file, "r", encoding="utf-8") as fin, open(output_file, "w", encoding="utf-8") as fout:
    for line in fin:
        line = line.strip()
        columns = line.split()
        fout.write('\t'.join(columns) + '\n')
```
