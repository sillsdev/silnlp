```python
def get_dict(filepath):
    result = {}
    with open(filepath, 'r', encoding='utf-8') as f:
        current_key = None
        for line in f:
            line = line.strip()
            if not line:
                continue

            if line.endswith(':'):
                current_key = line[:-1]

            elif current_key and ':' in line:
                word2_part, rest = line.split(':', 1)
                word2 = word2_part.strip()
                number_str = rest.split(',')[0].strip()
                number = float(number_str)
                result[current_key] = (word2, number)
    return result
```

```python
word_dict = get_dict('word_pairs.txt')
print(word_dict)
```

```python
def get_trg_words(filepath):
    trg_words = []
    with open(filepath, 'r', encoding='utf-8') as f:
        for line in f:
            line = line.split()
            if len(line) > 2 or len(line) == 0:
                continue
            trg_words.append(line)
    return trg_words
```

```python
trg_words = get_trg_words('src_words.txt')
print(trg_words)
```

```python
with open('word_pairs_sorted.txt', 'w', encoding='utf-8') as f:
    for word in trg_words:
        if word[0] in word_dict:
            f.write(f"{word[0]} {word[1]} {word_dict[word[0]][0]} {word_dict[word[0]][1]}\n")
```
