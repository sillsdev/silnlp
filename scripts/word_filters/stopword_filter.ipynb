For Target words filtering

```python
trg_words = {}
with open("lexicon.eflomal_clean.txt", 'r', encoding='utf-8') as file:
    for line in file:
        line = line.strip().split()
        if len(line) >= 3:
            key = line[0]
            value = line[1:]
            if key not in trg_words:
                trg_words[key] = []
            trg_words[key].append(value)
print(trg_words)
```

```python
with open('lexicon.eflomal_clean_filtered.txt', 'w') as file:
    for key in trg_words:
        for value in trg_words[key]:
            if value[0] in stopwords:
                continue
            file.write(f"{key} ")
            for v in value:
                file.write(f"{v} ")
            file.write("\n")
```

For Source words filtering

```python
src_words = []
with open('src_words.txt', 'r') as file:
    for line in file:
        parts = line.strip().split()
        if len(parts) >= 2:
            word, count = parts
            src_words.append((word, int(count)))

print(src_words)
```

```python
with open('filtered_src_words.txt', 'w') as file:
    for word, count in src_words:
        if word not in stopwords:
            file.write(f"{word} {count}\n")
```
